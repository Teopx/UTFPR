```python
#04
def encontra_palavras_que_comecam_com(lista_palavras, letra):
  palavras_encontradas = []
  for palavra in lista_palavras:
    if palavra[0] == letra:
      palavras_encontradas.append(palavra)
  return palavras_encontradas

text = ['abcd', 'abc', 'def', 'bla', 'xis', 'top', 'bot', 'ser', 'amn']

letra_procurada = input("Digite a letra inicial: ")

resultado = encontra_palavras_que_comecam_com(text, letra_procurada)

if resultado:
  print("Palavras que começam com", letra_procurada, ":", resultado)
else:
  print("Nenhuma palavra encontrada.")
```
