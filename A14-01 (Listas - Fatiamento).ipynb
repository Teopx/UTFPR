```python
#02
nums = [23, 22, 44, 17, 77, 55, 1, 65, 82, 2]

menor_nr = min(nums)

maior_nr = max(nums)

soma_numeros = sum(nums)

print("O menor número é:", menor_nr)
print("O maior número é:", maior_nr)
print("A soma dos números é:", soma_numeros)
```
