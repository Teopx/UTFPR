```python
#02
def contar_caracteres(texto):
    letras = 0
    digitos = 0
    especiais = 0

    for caractere in texto:
        if caractere.isalpha():
            letras += 1
        elif caractere.isdigit():
            digitos += 1
        else:
            especiais += 1

    return {'letras': letras, 'digitos': digitos, 'especiais': especiais}

if __name__ == "__main__":
    texto = input("Digite uma string: ")
    resultado = contar_caracteres(texto)
    print(f"'{texto}' contém:")
    print(f"- {resultado['letras']} letras")
    print(f"- {resultado['digitos']} dígitos")
    print(f"- {resultado['especiais']} caracteres especiais")
```

```python
#03
def calcular_soma_e_media(string):
    soma = 0
    contador = 0

    for caractere in string:
        if caractere.isdigit():
            soma += int(caractere)
            contador += 1

    if contador == 0:
        return None, None

    media = soma / contador
    return soma, media

string = input("Digite uma string: ")

resultado = calcular_soma_e_media(string)

if resultado:
    soma, media = resultado
    print("A soma  é:", soma)
    print("A média  é:", media)
else:
    print("A string não contém dígitos numéricos.")
```
