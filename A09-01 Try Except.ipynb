```python
def obter_numero_float():
    try:
        x = float(input("Digite um número decimal: "))
        return x
    except ValueError:
        raise TypeError("Apenas valores floats são permitidos!")

try:
    resultado = obter_numero_float()
    print("O número digitado foi:", resultado)
except TypeError as e:
    print(e)
```
