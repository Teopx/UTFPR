```python
#01
lista = [100, 200, 300, 400, 500]
invertida = list(reversed(lista))
print(invertida)
```

```python
#02
def calcular_quadrados(lista):
  quadrados = []
  for nr in lista:
    quadrados.append(nr ** 2)
  return quadrados

nr = [1, 2, 3, 4, 5, 6, 7]
resultado = calcular_quadrados(nr)
print(resultado)
```

```python
#03
def substituir_numero(lista, nr_subst):
   for i in range(len(lista)):
    if lista[i] == nr_subst:
      lista[i] = -1

numeros = [1, 2, 3, 2, 4, 2]
nr_subst = int(input("Digite o número a ser substituído: "))
substituir_numero(numeros, nr_subst)
print(numeros)
```

```python
#04
def remover_nr(lista, nr_remover):
    while nr_remover in lista:
       lista.remove(nr_remover)

numeros = [1, 2, 3, 2, 4, 2]
nr_remover = int(input("Digite o número a ser removido: "))
remover_nr(numeros, nr_remover)
print(numeros)
```
