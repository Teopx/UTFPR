```python
#02
T1 = (1, 9, 1, 6, 3, 4)
soma = sum(x for x in T1 if isinstance(x, (int, float)))
print(soma)
```

```python
#03
T = (50, 10, 60, 70, 50, 50)
valor = int(input("Digite o valor que deseja contar na tupla: "))
ocorrencias = T.count(valor)
print(f"O valor {valor} aparece {ocorrencias} vezes na tupla.")
```

```python
#05
L = [(), (), (15,), ('a','b'),('a','b','c'), ('d'), ()]
L = [t for t in L if t]
print(L)
```

```python
#06
dados = [('Pedro', '72.20'), ('Wilian', '69.10'), ('Andre','70.5')]
dados_ordenados = sorted(dados, key=lambda x: float(x[1]))
print(dados_ordenados)
```
