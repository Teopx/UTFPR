```python
nr = int(input('Digite um valor: '))
contador = 1

while contador <= nr:
  print(f'Valor do contador é {contador}')
  contador += 1
```

```python
nr = int(input('Digite um valor: '))
contador = 1

while contador <= nr:
  print(f'Valor do contador é {contador}')
  contador += 1
else:
  print(f'Sequência finalizada')
```

```python
p = 5
soma = 0
count = 0

while p > 0:
    count += 1
    f = int(input('Digite um número: '))
    soma = soma + f
    p -= 1

# Calcula a média após todas as iterações
media = soma / count
print('A média dos números digitados:', media)
```

```python
import math
n = 1

while n<=5:

  r = math.sqrt(n)
  print(f'A raiz quadrada de {n} é {r:.4f}')
  n += 1
```
