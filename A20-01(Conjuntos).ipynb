```python
#01
num_set = set([0, 1, 2, 3, 4, 5])

numero = input("Digite um número: ")


numero = int(numero)


if numero in num_set:
    print('Elementos:')
    print(num_set)
    num_set.remove(numero)
    print(f"Item {numero} removido.")
else:
    print(f"O número {numero} não está no conjunto.")



print(num_set)
```

```python
#02
x = set(['Cavalo', 'Ovelha', 'Cabra'])
y = set(['Ovelha', 'Cabra'])
z = set(['Ovelha'])


print("y é subconjunto de x:", y <= x)
print("x é subconjunto de y:", x <= y)
print("z é subconjunto de y:", z <= y)
print("z é subconjunto de x:", z <= x)
```
