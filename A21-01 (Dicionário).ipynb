```python
#02
D = {'w':112, 'x':12, 'y':99, 'z':60}
maior_valor = max(D.values())
minimo_valor = min(D.values())

print(f'O Maior valor é: {maior_valor}')
print(f'O Maior valor é: {minimo_valor}')
```

```python
#03
D = {'Brasil':['verde','amarelo','azul'], 'Italia':['branco','verde','vermelho'], 'Espanha':['amarelo','vermelho']}
total_chaves_valores = 0
for cores in D.values():
  total_chaves_valores += len(cores)

print(f'A quantidade de itens dentro das listas são: {total_chaves_valores}')
```

```python
#04
A = {'Pedro':92, 'Claudio':56, 'Marcia':88, 'Carlos':17, 'Marcos':8}
B = {'Valdir':70, 'Marcos':8, 'Claudio':56, 'Luiz':97, 'Pedro':12}
def encontrar_iguais(A, B):
  iguais = {}

  for chave in set(A.keys()) & set(B.keys()):
    if A[chave] == B[chave]:
      iguais[chave] = A[chave]
  return iguais

resultado = encontrar_iguais(A, B)
print(resultado)
```
